```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn  as sb
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
sales_data=pd.concat([train,test] , ignore_index = True)
sales_data.head()
```

```python
sales_data.isnull().sum()
```

```python
sales_data["Item_Outlet_Sales"] = sales_data["Item_Outlet_Sales"].fillna((sales_data["Item_Outlet_Sales"].mean()))
sales_data["Item_Weight"] = sales_data["Item_Weight"].fillna((sales_data["Item_Weight"].mean()))
```

```python
sales_data.isnull().sum()
```

```python
sales_data["Outlet_Size"] = sales_data["Outlet_Size"].fillna((sales_data["Outlet_Size"].mode()[0]))
```

```python
Zero_Item_Visibility =(sales_data['Item_Visibility']==0)
print (Zero_Item_Visibility.sum())
```

```python
sales_data['new_Item_Identifier']=sales_data['Item_Identifier'].apply(lambda x:x[0:2])
```

```python
sales_data.isnull().sum()
```

```python
sales_data['new_Item_Identifier']=sales_data['new_Item_Identifier'].replace({'FD':'food','NC':'non-consumable','DR':'drinks'})
```

```python
sales_data.head()
```

```python
sales_data['Item_Fat_Content']=sales_data['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular','low fat':'Low Fat'})
sales_data['Item_Fat_Content'].value_counts()
```

```python
sales_data.head()
```

```python
sales_data.drop(['Item_Outlet_Sales'],axis=1,inplace=True)
```

```python
sales_data.head()
```

```python
sales_data['Item_Fat_Content'].value_counts()
```

```python
sales_data.head()
```

```python
sales_data.drop(['Item_Identifier'],axis=1,inplace=True)
```

```python
sales_data.head()
```

```python
sales_data.drop(['Outlet_Identifier'],axis=1,inplace=True)
```

```python
sales_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_x_1=LabelEncoder()
```

```python
sales_data['new_Item_Identifier'] = labelencoder_x_1.fit_transform(sales_data['new_Item_Identifier'])
```

```python
sales_data['new_Item_Identifier'].value_counts()
```

```python
sales_data['Item_Fat_Content'] = labelencoder_x_1.fit_transform(sales_data['Item_Fat_Content'])
sales_data['Item_Type'] = labelencoder_x_1.fit_transform(sales_data['Item_Type'])
sales_data['Outlet_Location_Type'] = labelencoder_x_1.fit_transform(sales_data['Outlet_Location_Type'])
sales_data['Outlet_Size'] = labelencoder_x_1.fit_transform(sales_data['Outlet_Size'])
```

```python
sales_data['Outlet_Type'] = labelencoder_x_1.fit_transform(sales_data['Outlet_Type'])
```

```python
sales_data.head()
```

```python
sales_data.loc[Zero_Item_Visibility,'Item_Visibility']=sales_data["Item_Visibility"].mean() 
```

```python
x=sales_data
```

```python
y=train['Item_Outlet_Sales']
```

```python
x_train = x[:len(train)]
x_test = x[len(train):]
```

```python
x_train.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
xtrain, xtest, ytrain, ytest = train_test_split(x_train, y,test_size = 0.1, random_state = 0)
```

```python
sc_x=StandardScaler()
xtrain= sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)
```

```python
#importing keras
import keras
from keras.models import Sequential
from keras.layers import Dense

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(10, activation='relu', input_dim = 10))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(1))

model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
model.fit(xtrain,ytrain,batch_size=10,nb_epoch=200)
```

```python
y_pred=model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(ytest, y_pred))
```

```python
rmse = np.sqrt(mean_squared_error(ytest, y_pred))
rmse
```
